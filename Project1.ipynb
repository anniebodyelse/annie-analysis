```python
def create_animated_choropleth(dataframe,value_col,title, color_scale=px.colors.sequential.Plasma):
    """
    Create an animated choropleth map using Plotly.

    Parameters:
    - dataframe (pd.DataFrame): The source dataframe.
    - locations_col (str): Column name for country names or codes.
    - frame_col (str): Column name for animation frames (e.g., 'Year').
    - value_col (str): Column name for the metric to visualize.
    - title (str): Title for the map. Defaults to None.
    - location_mode (str): Either 'country names' or 'ISO-3' based on dataframe data. Defaults to 'country names'.
    - color_scale (list or str): Color scale for the visualization. Defaults to px.colors.sequential.Plasma.

    Returns:
    - plotly.graph_objs._figure.Figure: Animated choropleth map.
    """

    fig = px.choropleth(dataframe, 
                        locations='Entity', 
                        locationmode='country names', 
                        color=value_col,
                        hover_name='Entity', 
                        animation_frame='Year', 
                        title=title if title else f'Animated Choropleth Map of {value_col}',
                        color_continuous_scale=color_scale
                       )
    fig.update_geos(
        showcoastlines = True,
        coastlinecolor = "Black",
        showland = True,
        landcolor = "white",
        showcountries = True,
        showocean = True,
        oceancolor = "LightBlue")
    fig.show()

# Example Usage:
# create_animated_choropleth(df, 'Entity', 'Year', 'Access to electricity (% of population)')
```
